```python
import control
import matplotlib.pyplot as plt
def bode(Quad):
    control.bode_plot(Quad, dB=True)
    plt.show()


T = 8.769559 # kg/m^2 #thrust
l = 1  # m
m = 0.6  # kg
g = 9.81  # m/s^2

Quad= control.tf([1], [T, 0, -m * g * l])

def pid(Kp, Ki, Kd):
    num = [Kd, Kp, Ki]
    den = [1, 0]
    return control.tf(num, den)

Kp = 1.0
Ki = 0.1
Kd = 0.01

C = pid(Kp, Ki, Kd)

closed_loop = control.feedback(C*Quad, 1)

bode(closed_loop)

damping = control.damp(closed_loop)

Data = control.step_info(closed_loop)

Data 
```
